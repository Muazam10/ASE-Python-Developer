### **Today we will Learn "Set" and "Dictionary"**

##### **1.Set**

```python
# Create a set
numbers = {1, 2, 3, 4}
print(numbers)
# Add an element to the set
numbers.add(5) # practice: remove, discard, pop, clear
print(numbers)
```

```python
# Duplicate values auto removed in sets
data = {1, 2, 2, 3, 3, 4}
print(data)

# Check value exit or not
users = {"Ali", "Ahmed", "Sara"}

print("Ali" in users)
print("Bilal" in users)
```

```python
# union of sets
a = {1, 2, 3}
b = {3, 4, 5}
c = a.union(b)
print(c)  # Output: {1, 2, 3, 4,
```

```python
# intersection of sets
x = {10, 20, 30}
y = {20, 40, 50}

print(x.intersection(y))
```

##### 2. Dictionary

```python
# Creating a Dictionary 
person = {
    "name": "Ali",
    "age": 30,
    "city": "Lahore"
}
print(person) 
```

```python
print(person["name"])
print(person.get("age"))
```

```python
# add a new key-value pair
person["profession"] = "Engineer"
print(person)
# update an existing value
person["age"] = 31
print(person)
# delete data from dictionary
del person["city"]
print(person)
```

```python
# loop through dictionary
for key, value in person.items():
    print(f"{key}: {value}")

# check if key exists
if "name" in person:
    print("Name exists in the dictionary.")
```

```python
# Set  & Dictionary practice with examples
emails = ["a@gmail.com", "b@gmail.com", "a@gmail.com"]

unique_emails = set(emails)
print(unique_emails)

users = {
    1: "Ali",
    2: "Ahmed",
    3: "Sara"
}

print(users[1])
```
