### **Today we will Learn "Set" and "Dictionary"**

##### **1.Set**

```python
# Create a set
numbers = {1, 2, 3, 4}
print(numbers)
# Add an element to the set
numbers.add(5) # practice: remove, discard, pop, clear
print(numbers)
```

```python
# Duplicate values auto removed in sets
data = {1, 2, 2, 3, 3, 4}
print(data)

# Check value exit or not
users = {"Ali", "Ahmed", "Sara"}

print("Ali" in users)
print("Bilal" in users)
```

```python
# union of sets
a = {1, 2, 3}
b = {3, 4, 5}
c = a.union(b)
print(c)  # Output: {1, 2, 3, 4,
```

```python
# intersection of sets
x = {10, 20, 30}
y = {20, 40, 50}

print(x.intersection(y))
```
