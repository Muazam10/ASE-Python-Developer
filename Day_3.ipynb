# **Loops**

### ***for loop***

```python
for i in range(1, 6):
    print(i)
```

```python
# Print whether numbers from 1 to 10 are even or odd
for i in range(1, 11):
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")
```

```python
# Print multiplication table of a given number
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n*i)
```
