# **Loops**

### ***for loop***

```python
for i in range(1, 21):
    print(i)
```

```python
# Print whether numbers from 1 to 10 are even or odd

for i in range(1, 11):
    if i % 2 == 0:
        print(i, "is Even")
    else:
        print(i, "is Odd")
```

```python
# Print multiplication table of a given number
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n*i)
```

```python
# Using while loop for print reverse numbers from 10 to 1
count = 10  
while count >= 1:
    print(count)
    count -= 1
# using for loop for print reverse numbers from 10 to 1
for i in range(10, 0, -1):
    print(i)
```
