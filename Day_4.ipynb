```python
# Student Management System
# Using List & Tuple (Professional Approach)

students = []   # list of tuples


def calculate_grade(marks):
    if marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"


def add_student():
    student_id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    marks = float(input("Enter Marks: "))

    grade = calculate_grade(marks)

    student = (student_id, name, marks, grade)  # tuple
    students.append(student)

    print("âœ… Student added successfully\n")


def view_students():
    if not students:
        print("No students found\n")
        return

    print("ID | Name | Marks | Grade")
    print("-" * 30)

    for s in students:
        print(s[0], "|", s[1], "|", s[2], "|", s[3])

    print()


def find_topper():
    if not students:
        print("No data available\n")
        return

    topper = max(students, key=lambda s: s[2])
    print("ğŸ† Topper:", topper[1], "with", topper[2], "marks\n")


def class_average():
    if not students:
        print("No data available\n")
        return

    total = sum(s[2] for s in students)
    average = total / len(students)

    print("ğŸ“Š Class Average:", round(average, 2), "\n")


# Main Menu
while True:
    print("1. Add Student")
    print("2. View Students")
    print("3. Find Topper")
    print("4. Class Average")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        find_topper()
    elif choice == "4":
        class_average()
    elif choice == "5":
        print("Exiting program...")
        break
    else:
        print("Invalid choice\n")
```
