```python
students = []

while True:
    print("\nStudent Management System")
    print("1. Add student")
    print("2. View students")
    print("3. Show topper")
    print("4. Show class average")
    print("5. Exit")

    choice = input("Choose an option: ")

    # Add student
    if choice == "1":
        sid = int(input("Student ID: "))
        name = input("Student name: ")
        marks = float(input("Marks: "))

        if marks >= 80:
            grade = "A"
        elif marks >= 70:
            grade = "B"
        elif marks >= 60:
            grade = "C"
        else:
            grade = "Fail"

        students.append((sid, name, marks, grade))
        print("Student added successfully.")

    # View students
    elif choice == "2":
        if not students:
            print("No students added yet.")
        else:
            print("\nID | Name | Marks | Grade")
            print("-" * 30)
            for s in students:
                print(s[0], "|", s[1], "|", s[2], "|", s[3])

    # Find topper
    elif choice == "3":
        if not students:
            print("No data available.")
        else:
            topper = max(students, key=lambda s: s[2])
            print("Topper is:", topper[1])
            print("Marks:", topper[2])

    # Class average
    elif choice == "4":
        if not students:
            print("No data available.")
        else:
            total = sum(s[2] for s in students)
            avg = total / len(students)
            print("Class average marks:", round(avg, 2))

    # Exit
    elif choice == "5":
        print("Goodbye ðŸ‘‹")
        break

    else:
        print("Invalid option, try again.")
```
