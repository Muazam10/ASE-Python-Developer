# **Today we will Learn "List" and "Tuple"**

# **1. List**

#### **List = ordered, mutable (changeable) collection**

```python
# Creating lists with different data types
numbers = [10, 20, 30, 40]
names = ["Ali", "Ahmed", "Sara"]
mixed = [1, "Python", True, 3.5]
print("Numbers List:", numbers)
print("Names List:", names) 
print("Mixed List:", mixed)
```

```python
# Accessing elements using indexing 
print(numbers[0])   # 10
print(numbers[-1])  # 40
print(names[1])     # Ahmed
print(mixed[2])     # True
```

```python
# Modifying list elements
numbers[1] = 25 # Change 20 to 25
```

```python
# looping through a list
for num in numbers: # Print each number by sequentially accessing elements
    print(num)
```

## **2. Tuple**

#### **Tuple = ordered, immutable (cannot change)**

```python
# Create tuples with different data types
single = (5,) # comma is necessary for single-element tuples
point = (10, 20, 30, 40)
person = ("Ali", 30, True) 
print("Single Element Tuple:", single)
print("Point Tuple:", point)
print("Person Tuple:", person)
```

```python
print(point[1])   # 20
print(point[-2])  # 30
print(person[0])  # Ali
```

```python
# Using Loop to access tuple elements
for value in point:
    print(value)
```

# ***Real Life Example for practice***

```python
# Real-world application: Student Management System
# Each student record: (ID, Name, Marks, Grade)
students = []

while True:
    print("\nStudent Management System")
    print("1. Add student")
    print("2. View students")
    print("3. Show topper")
    print("4. Show class average")
    print("5. Exit")

    choice = input("Choose an option: ")

    # Add student
    if choice == "1":
        sid = int(input("Student ID: "))
        name = input("Student name: ")
        marks = float(input("Marks: "))

        if marks >= 80:
            grade = "A"
        elif marks >= 70:
            grade = "B"
        elif marks >= 60:
            grade = "C"
        else:
            grade = "Fail"

        students.append((sid, name, marks, grade))
        print("Student added successfully.")

    # View students
    elif choice == "2":
        if not students:
            print("No students added yet.")
        else:
            print("\nID | Name | Marks | Grade")
            print("-" * 30)
            for s in students:
                print(s[0], "|", s[1], "|", s[2], "|", s[3])

    # Find topper
    elif choice == "3":
        if not students:
            print("No data available.")
        else:
            topper = max(students, key=lambda s: s[2])
            print("Topper is:", topper[1])
            print("Marks:", topper[2])

    # Class average
    elif choice == "4":
        if not students:
            print("No data available.")
        else:
            total = sum(s[2] for s in students)
            avg = total / len(students)
            print("Class average marks:", round(avg, 2))

    # Exit
    elif choice == "5":
        print("Goodbye ðŸ‘‹")
        break

    else:
        print("Invalid option, try again.")
```
