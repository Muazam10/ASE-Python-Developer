# **Smart Student Result & Attendance Analyzer**

```python
# Smart Student Result & Attendance Analyzer
# Week 1 Project: Python Basics, Variables, Loops, Conditions

# Counters (summary ke liye)
total_students = 0
passed_students = 0
failed_students = 0
grade_a_students = 0
highest_marks = -1
average_marks = 0

# Infinite loop (jab tak user exit na kare)
while True:
    name = input("\nEnter student name (or type 'exit' to stop): ")

    # Exit condition
    if name.lower() == "exit":
        break  # loop end

    # Marks input
    marks = int(input("Enter marks (0-100): "))
    attendance = int(input("Enter attendance percentage (0-100): "))

    # Validation check
    if marks < 0 or marks > 100 or attendance < 0 or attendance > 100:
        print("Invalid input, student skipped")
        continue  # next student

    # Valid student count
    total_students += 1

    # Attendance condition (highest priority)
    if attendance < 75:
        print(f"Stuck off (Short Attendance) : {attendance}%")
        failed_students += 1

    else:
        # Marks based result
        if marks >= 85:
            print("PASS (Grade A)")
            passed_students += 1
            grade_a_students += 1

        elif marks >= 70:
            print("PASS (Grade B)")
            passed_students += 1

        elif marks >= 50:
            print("PASS (Grade C)")
            passed_students += 1

        else:
            print("FAIL")
            failed_students += 1
mark_average = (average_marks * (total_students - 1) + marks) / total_students
average_marks = mark_average
# Update highest marks
if marks > highest_marks:
   highest_marks = marks   
# Current summary   
print("\n------ SUMMARY ------")
print("Total Students:", total_students)
print("Passed Students:", passed_students)
print("Failed Students:", failed_students)
print("Grade A Students:", grade_a_students)
print("Highest Marks so far:", highest_marks)
print("Average Marks so far:", round(average_marks, 2))
```
